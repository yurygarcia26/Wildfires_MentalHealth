```python
wildfire_name = 'Tubbs'
output_name1  = '2.TubbsFire_wildfire_words.csv'
```

```python
# Load the CSV file into a Pandas DataFrame
df = pd.read_csv(os.path.join(files_path, output_name1),dtype=dtypes)
print(df.shape)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have a DataFrame called df with a 'tweet_created_at' column

# Convert 'tweet_created_at' column to datetime format
df['tweet_created_at'] = pd.to_datetime(df['tweet_created_at'])

# Extract the hour and day from the 'tweet_created_at' column
df['hour'] = df['tweet_created_at'].dt.hour
df['day']  = df['tweet_created_at'].dt.date

# Group by day and hour and calculate the count of tweets
tweet_counts = df.groupby(['day', 'hour']).size().unstack(fill_value=0)

# Create a heatmap from the tweet counts
plt.figure(figsize=(10, 6))
sns.heatmap(tweet_counts, cmap='YlGnBu', cbar=True)
plt.title('')
plt.xlabel('Hour')
plt.ylabel('Day')



# Save the figure as a PDF
#figure_path = '/content/drive/MyDrive/Mental_Health_Wildfire/Twitter_Data/Tubbs_Codes/Figures/'
#figure_name = wildfire_name + '_tweets_per_houre'
#output_file = os.path.join(figure_path, figure_name)
#plt.savefig(output_file, format='pdf')


plt.show()

```
